```python
import numpy as np
import math
import matplotlib.pyplot as plt
import os
import sys
import io
import re
import collections
import random
```

```python
class ZipfsLaw:
    """
        This experiment is shown that any chosen corpora which is suffciently large to
        be observed from the randomized or combined version to the corpus with a
        complex structure such as novels can show the Zipfian behaviour. The
        frequency of the occurrence of the word is approximately the inverse power law
        function of its rank. It is shown that every corpus has the characteristics, one
        of the characteristics is the relationship between its token size and its
        vocabulary size. If the relationship between the token size and the vocabulary
        size of a corpus can be obtained, it is possible to classify the corpus and group
        with respect to its author or its type of literature, since the token-type data
        points is grouped to create its characteristic curve.
    """
    def __init__(self, file):
        self.file = file
        self.file_without_stop = None
    def tokenization_corpus(self):
        """
            Tokenize the corpus.
            Get rid of the punctuation marks.
            Set every word to lowercase.
        """
        try:
            old_file = self.file
            with open(self.file, mode = "r", encoding="utf-8") as file:
                for line in file:
                    file_write = io.open("C:\\Users\\lenovo\\The Corpora\\Three Authors\\Charles Dickens\\temp.txt", mode = "a", encoding="utf-8")
                    for ch in ['\\','`','*','_','{','}','[',']','(',')','>','#','+','-','.','!','$','\'',',','?',':',';','!']:
                        if ch in line:
                            line = line.replace(ch," ")
                    line = line.lower()
                    file_write.write(line)
                    file_write.close()
                file.close()
            print(file.closed)
            os.unlink(self.file)
            os.rename("C:\\Users\\lenovo\\The Corpora\\Three Authors\\Charles Dickens\\temp.txt",old_file)
            self.file = old_file
        except FileNotFoundError:
            print("File is not found !")
            
    def stopword_corpus(self):
        """
            Get rid of the stop words.
        """
        try:
            old_file = self.file
            new_file = old_file[:-4] + "_without_stopword.txt"
            with open(self.file, mode = "r", encoding="utf-8") as file:
                for line in file:
                    file_write = io.open(new_file, mode = "a", encoding="utf-8")
                    for word in ["not", "a", "about", "above", "after", "again", "against", "ain", "all", "am", "an", "and", "any", "are", "aren", "arent", "as", "at", "be", "because", "been", "before", "being", "below", "between", "both", "but", "by", "can", "couldn", "couldnt", "d", "did", "didn", "didnt", "do", "does", "doesn", "doesnt", "doing", "don", "dont", "down", "during", "each", "few", "for", "from", "further", "had", "hadn", "hadn't", "has", "hasn", "hasnt", "have", "haven", "havent", "having", "he", "her", "here", "hers", "herself", "him", "himself", "his", "how", "i", "if", "in", "into", "is", "isn", "isnt", "it", "its", "its", "itself", "just", "ll", "m", "ma", "me", "mightn", "mightnt", "more", "most", "mustn", "mustnt", "my", "myself", "needn", "neednt", "now", "o", "of", "off", "on", "once", "only", "or", "other", "our", "ours", "ourselves", "out", "over", "own", "re", "s", "same", "shan", "shant", "she", "shes", "should", "shouldve", "shouldn", "shouldnt", "so", "some", "such", "t", "than", "that", "thatll", "the", "their", "theirs", "them", "themselves", "then", "there", "these", "they", "this", "those", "through", "to", "too", "under", "until", "up", "ve", "very", "was", "wasn", "wasnt", "we", "were", "weren", "werent", "what", "when", "where", "which", "while", "who", "whom", "why", "will", "with", "won", "wont", "wouldn", "wouldnt", "y", "you", "youd", "youll", "youre", "youve", "your", "yours", "yourself", "yourselves", "could", "hed" , "hes", "heres", "hows" , "im", "ive", "lets", "ought", "thats", "theres", "theyd", "theyll", "theyre", "theyve", "wed", "were", "weve", "whats", "whens", "wheres", "whos", "whys", "would"]:   
                        pattern = "\\b" + word + "\\b"
                        line = re.sub(pattern, "" , line) 
                    file_write.write(line)
                    file_write.close()
                file.close()
            print(file.closed)
            self.file_without_stop = new_file
        except FileNotFoundError:
            print("File is not found !")
            
    def get_occurences(self):
        """
            Get the occurences of words and write to a separate file.
        """
        try:
            word = []
            word_without_stop = []
            occur = self.file[:-4] + "_occurences.txt"
            occur_stop = self.file[:-4] + "_without_stopword_occurences.txt"
            with open(self.file, mode = "r", encoding="utf-8") as file:
                for line in file:
                    word.extend(line.split())
                file.close()
            print(file.closed)
            with open(self.file_without_stop, mode = "r", encoding="utf-8") as filestop:
                for line in filestop:
                    word_without_stop.extend(line.split())
                filestop.close()
            print(filestop.closed)
            count = collections.Counter(word)
            count_stop = collections.Counter(word_without_stop)
            count_tuple = count.most_common()
            count_stop_tuple = count_stop.most_common()
            with open(occur, mode = "a", encoding="utf-8") as fileoccur:
                for element in count_tuple:
                    fileoccur.write(element[0] + " : " + str(element[1]) + "\n")
                fileoccur.close()
            print(fileoccur.closed)
            with open(occur_stop, mode = "a", encoding="utf-8") as fileoccurstop:
                for element in count_stop_tuple:
                    fileoccurstop.write(element[0] + " : " + str(element[1]) + "\n")
                fileoccurstop.close()
            print(fileoccurstop.closed)
            return count, count_stop
        except FileNotFoundError:
            print("File is not found !")
    
    def random_get_occurence(self):
        """
            Get the occurences of words of randomly created corpus and write to a separate file.
        """
        try:
            word = []
            occur = self.file[:-4] + "_occurence.txt"
            with open(self.file, mode = "r", encoding="utf-8") as file:
                for line in file:
                    word.extend(line.split())
                file.close()
            print(file.closed)
            count = collections.Counter(word)
            count_tuple = count.most_common()
            with open(occur, mode = "a", encoding="utf-8") as fileoccur:
                for element in count_tuple:
                    fileoccur.write(element[0] + " : " + str(element[1]) + "\n")
                fileoccur.close()
            print(fileoccur.closed)
            return count
        except FileNotFoundError:
            print("File is not found !") 
            
    @staticmethod
    
    def plot_zipfslaw(first_count, second_count, third_count, author, log_log = False):
        """
            Plot the relationship between the frequency and the rank to see Zipfian behavior 
            on the combined corpora with respect to its author or type of literature.
        """
        big_corpora = first_count + second_count + third_count
        y_freq = []
        big_tuple = big_corpora.most_common()
        if log_log == True:
            for element in big_tuple:
                y_freq.append(math.log(element[1]))
            x_new = [math.log(i) for i in range(1, len(y_freq) + 1)]
        else:
            for element in big_tuple:
                y_freq.append(element[1]) 
            x_new = [i for i in range(1, len(y_freq) + 1)]
        plt.plot(x_new, y_freq, 'bo', markersize=1)
        if log_log == True:
            plt.title(author + " Log-Log Graph", fontsize = 24)
        else:
            plt.title(author + " Normal Graph", fontsize = 24)
        plt.ylabel('Frequency')
        plt.xlabel('Rank')
        plt.xticks(rotation = 45)
        plt.show()
    
    
    @staticmethod
    def plot_seperate(first_count, second_count, third_count, author, books, stopword = False):
        """
            Plot the relationship between the frequency and the rank to see Zipfian behavior 
            on the each corpora with respect to its author or type of literature.
        """
        plt.figure(figsize=(21, 7))
        y_freq = []
        big_tuple = first_count.most_common()
        for element in big_tuple:
            y_freq.append(math.log(element[1]))
        x_new = [math.log(i) for i in range(1, len(y_freq) + 1)]
        plt.subplot(131)
        plt.plot(x_new, y_freq, 'bo', markersize=1)
        plt.title(books[0], fontsize=18)
        plt.ylabel('Frequency', fontsize=15)
        plt.xlabel('Rank', fontsize=15)
        plt.xticks(rotation = 45)
        y_freq = []
        big_tuple = second_count.most_common()
        for element in big_tuple:
            y_freq.append(math.log(element[1]))
        x_new = [math.log(i) for i in range(1, len(y_freq) + 1)]
        plt.subplot(132)
        plt.plot(x_new, y_freq, 'bo', markersize=1)
        plt.title(books[1], fontsize=18)
        plt.ylabel('Frequency', fontsize=15)
        plt.xlabel('Rank', fontsize=15)
        plt.xticks(rotation = 45)
        y_freq = []
        big_tuple = third_count.most_common()
        for element in big_tuple:
            y_freq.append(math.log(element[1]))
        x_new = [math.log(i) for i in range(1, len(y_freq) + 1)]
        plt.subplot(133)
        plt.plot(x_new, y_freq, 'bo', markersize=1)
        plt.title(books[2], fontsize=18)
        plt.ylabel('Frequency', fontsize=15)
        plt.xlabel('Rank', fontsize=15)
        plt.xticks(rotation = 45)
        if stopword == True:
            plt.suptitle(author + " Log-Log Graph without Stopwords", y=1.05, fontsize=24)
        else:
            plt.suptitle(author + " Log-Log Graph", y=1.05, fontsize=24)
        plt.show()
    
    @staticmethod
    def token_type_relation(author, first_book, second_book, third_book, stopword = False):
        """
            Plot the relationship between the token size and the vocabulary size to see a behavior 
            on the combined corpora with respect to its author or type of literature.
        """
        try:
            plt.figure(figsize=(20, 10))
            author_name = author
            x_token_size = []
            y_vocab_size = []
            x_token_sizelog = []
            y_vocab_sizelog = []
            token_size = 0
            vocab_size = 0
            token = []
            if stopword == False:
                new_file = author + "_big_corpora.txt"
                with open(first_book.file, mode = "r", encoding="utf-8") as first_file:
                    for line in first_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    first_file.close()
                print(first_file.closed)
                with open(second_book.file, mode = "r", encoding="utf-8") as second_file:
                    for line in second_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    second_file.close()
                print(second_file.closed)
                with open(third_book.file, mode = "r", encoding="utf-8") as third_file:
                    for line in third_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    third_file.close()
                print(third_file.closed)
                with open(new_file, mode = "r", encoding="utf-8") as big_file:
                    for line in big_file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                                token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_size.append(token_size)
                                y_vocab_size.append(vocab_size)
                                x_token_sizelog.append(math.log(token_size))
                                y_vocab_sizelog.append(math.log(vocab_size))
                    big_file.close()
                print(big_file.closed)
                plt.subplot(1, 2, 1)
                plt.plot(x_token_size, y_vocab_size, 'bo', markersize=1)
                plt.title(author_name + " Normal Graph", fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 2, 2)
                plt.plot(x_token_sizelog, y_vocab_sizelog, 'bo', markersize=1)
                plt.title(author_name + " Log-Log Graph", fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.suptitle(author_name + " Graph", y=1.05, fontsize=24)
                plt.show()
            else:
                new_file = author + "_big_corpora_without_stopword.txt"
                with open(first_book.file_without_stop, mode = "r", encoding="utf-8") as first_file:
                    for line in first_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    first_file.close()
                print(first_file.closed)
                with open(second_book.file_without_stop, mode = "r", encoding="utf-8") as second_file:
                    for line in second_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    second_file.close()
                print(second_file.closed)
                with open(third_book.file_without_stop, mode = "r", encoding="utf-8") as third_file:
                    for line in third_file:
                        file_write = io.open(new_file, mode = "a", encoding="utf-8")
                        file_write.write(line)
                        file_write.close()
                    third_file.close()
                print(third_file.closed)
                with open(new_file, mode = "r", encoding="utf-8") as big_file:
                    for line in big_file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                                token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_size.append(token_size)
                                y_vocab_size.append(vocab_size)
                                x_token_sizelog.append(math.log(token_size))
                                y_vocab_sizelog.append(math.log(vocab_size))
                    big_file.close()
                print(big_file.closed)
                plt.subplot(1, 2, 1)
                plt.plot(x_token_size, y_vocab_size, 'bo', markersize=1)
                plt.title(author_name + " Normal Graph", fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 2, 2)
                plt.plot(x_token_sizelog, y_vocab_sizelog, 'bo', markersize=1)
                plt.title(author_name + " Log-Log Graph", fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.suptitle(author_name + " Graph without Stopwords", y=1.05, fontsize=24)
                plt.show()
            return x_token_sizelog, y_vocab_sizelog
        except FileNotFoundError:
            print("File is not found !")
    
    @staticmethod
    def token_type_separate(author, first_book, second_book, third_book, books, stopword = False):
        """
            Plot the relationship between the token size and the vocabulary size to see a behavior 
            on the each separate corpora with respect to its author or type of literature.
        """
        try:
            author_name = author
            x_token_sizelog_first = []
            y_vocab_sizelog_first = []
            x_token_sizelog_second = []
            y_vocab_sizelog_second = []
            x_token_sizelog_third = []
            y_vocab_sizelog_third = []
            plt.figure(figsize=(30, 10))
            if stopword == False:
                token_size = 0
                vocab_size = 0
                token = []
                with open(first_book.file, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_first.append(math.log(token_size))
                                y_vocab_sizelog_first.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                token_size = 0
                vocab_size = 0
                token = []
                with open(second_book.file, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_second.append(math.log(token_size))
                                y_vocab_sizelog_second.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                token_size = 0
                vocab_size = 0
                token = []
                with open(third_book.file, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_third.append(math.log(token_size))
                                y_vocab_sizelog_third.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                plt.subplot(1, 3, 1)
                plt.plot(x_token_sizelog_first, y_vocab_sizelog_first, 'bo', markersize=1)
                plt.title(books[0], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 3, 2)
                plt.plot(x_token_sizelog_second, y_vocab_sizelog_second, 'bo', markersize=1)
                plt.title(books[1], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 3, 3)
                plt.plot(x_token_sizelog_third, y_vocab_sizelog_third, 'bo', markersize=1)
                plt.title(books[2], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.suptitle(author_name + " Graph", y=1.05, fontsize=24)
                plt.show()
            else:
                token_size = 0
                vocab_size = 0
                token = []
                with open(first_book.file_without_stop, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_first.append(math.log(token_size))
                                y_vocab_sizelog_first.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                token_size = 0
                vocab_size = 0
                token = []
                with open(second_book.file_without_stop, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_second.append(math.log(token_size))
                                y_vocab_sizelog_second.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                token_size = 0
                vocab_size = 0
                token = []
                with open(third_book.file_without_stop, mode = "r", encoding="utf-8") as file:
                    for line in file:
                        words = line.split()
                        for word in words:
                            if word in token:
                                pass
                            else:
                                vocab_size += 1
                            token.append(word)
                            token_size += 1
                            if token_size % 5000 == 0:
                                x_token_sizelog_third.append(math.log(token_size))
                                y_vocab_sizelog_third.append(math.log(vocab_size))
                    file.close()
                print(file.closed)
                plt.subplot(1, 3, 1)
                plt.plot(x_token_sizelog_first, y_vocab_sizelog_first, 'bo', markersize=1)
                plt.title(books[0], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 3, 2)
                plt.plot(x_token_sizelog_second, y_vocab_sizelog_second, 'bo', markersize=1)
                plt.title(books[1], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.subplot(1, 3, 3)
                plt.plot(x_token_sizelog_third, y_vocab_sizelog_third, 'bo', markersize=1)
                plt.title(books[2], fontsize = 18)
                plt.ylabel('Vocabulary Size', fontsize=15)
                plt.xlabel('Token Size', fontsize=15)
                plt.xticks(rotation = 45)
                plt.suptitle(author_name + " Graph", y=1.05, fontsize=24)
                plt.show()
            return x_token_sizelog_first, y_vocab_sizelog_first, x_token_sizelog_second, y_vocab_sizelog_second, x_token_sizelog_third, y_vocab_sizelog_third
        except FileNotFoundError:
            print("File is not found !")
            
    @staticmethod
    def token_type_combine(authors, first_author, second_author, third_author):
        """
            Combine the plots of the relationship between the token size and the vocabulary size of
            the each separate corpora with respect to its author or type of literature.
        """
        plt.title("Combined Plots of Seperate Books", fontsize=24)
        plt.plot(first_author[0], first_author[1], "bo", label=authors[0], markersize=1)
        plt.plot(first_author[2], first_author[3], "bo", markersize=1)
        plt.plot(first_author[4], first_author[5], "bo", markersize=1)
        plt.plot(second_author[0], second_author[1], "ro", label=authors[1], markersize=1)
        plt.plot(second_author[2], second_author[3], "ro", markersize=1)
        plt.plot(second_author[4], second_author[5], "ro", markersize=1)
        plt.plot(third_author[0], third_author[1], "go", label=authors[2], markersize=1)
        plt.plot(third_author[2], third_author[3], "go", markersize=1)
        plt.plot(third_author[4], third_author[5], "go", markersize=1)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.legend(loc="lower right")
        plt.xticks(rotation = 45)
        plt.show()
        
    @staticmethod
    def token_type_combine_bigcorpora(authors, first_author, second_author, third_author):
        """
            Combine the plots of the relationship between the token size and the vocabulary size of
            the each combined corpora with respect to its author or type of literature.
        """
        plt.title("Combined Plots of Seperate Authors", fontsize=24)
        plt.plot(first_author[0], first_author[1], "-bo", label=authors[0], markersize=1)
        plt.plot(second_author[0], second_author[1], "-go", label=authors[1], markersize=1)
        plt.plot(third_author[0], third_author[1], "-ro", label=authors[2], markersize=1)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.legend(loc="lower right")
        plt.xticks(rotation = 45)
        plt.show()
        
    @staticmethod
    def find_slope(token_size, vocab_size):
        """
            Find the slope and the bias of best fitting lines for the datapoints of the token size 
            and the vocabulary size of the each combined or separate corpora with respect to 
            its author or type of literature.
        """
        np_token = np.asarray(token_size)
        np_vocab = np.asarray(vocab_size)
        token_dist = np_token - np.mean(np_token)
        vocab_dist = np_vocab - np.mean(np_vocab)
        slope = np.matmul(token_dist, vocab_dist.transpose()) / np.matmul(token_dist, token_dist.transpose())
        bias = np.mean(np_vocab) - slope * np.mean(np_token)
        return slope, bias
    
    @staticmethod
    def plot_predicted_actual(model, actual, titles, suptitle, stopword = False):
        """
            Plot the best fitting lines and emprical lines for the datapoints of the token size and the vocabulary size of
            the each combined or separate corpora with respect to its author or type of literature.
        """
        plt.figure(figsize=(30, 10))
        st_name = suptitle
        plt.subplot(1, 3, 1)
        np_predicted = model[0][0] * np.asarray(actual[0]) + model[0][1]
        plt.plot(actual[0], np_predicted, 'b', label="Predicted")
        plt.plot(actual[0], actual[1], 'r', label="Actual")
        plt.title(titles[0], fontsize = 18)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.legend(loc="lower right", fontsize=15)
        plt.xticks(rotation = 45)
        plt.subplot(1, 3, 2)
        np_predicted = model[1][0] * np.asarray(actual[2]) + model[1][1]
        plt.plot(actual[2], np_predicted, 'b', label="Predicted")
        plt.plot(actual[2], actual[3], 'r', label="Actual")  
        plt.title(titles[1], fontsize = 18)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.legend(loc="lower right", fontsize=15)
        plt.xticks(rotation = 45)
        plt.subplot(1, 3, 3)
        np_predicted = model[2][0] * np.asarray(actual[4]) + model[2][1]
        plt.plot(actual[4], np_predicted, 'b', label="Predicted")
        plt.plot(actual[4], actual[5], 'r', label="Actual")
        plt.title(titles[2], fontsize = 18)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.legend(loc="lower right", fontsize=15)
        plt.xticks(rotation = 45)
        if stopword == True:
            plt.suptitle(st_name + " Predicted and Actual Log-Log Graph without Stopwords", y=1.05, fontsize=24)
        else:
            plt.suptitle(st_name + " Predicted and Actual Log-Log Graph", y=1.05, fontsize=24)
        plt.show()
    
    def random_corpus_generator(self):
        """
            Generate a random corpus to see that whether the corpus shows the Zipfian behavior.
            Plot the results in the normal and the log-log format.
        """
        alphabets = [i for i in range(97, 123)]
        alphabets.append(32)
        plt.figure(figsize=(20, 10))
        count = 0
        with open(self.file, mode = "a", encoding="utf-8") as random_file:
            for i in range(4000000):
                random.shuffle(alphabets)
                random_index = np.random.randint(0, len(alphabets) - 1) 
                if random_index == 32:
                    count += 1
                random_alphabet = alphabets[random_index]
                if i == 0 and random_alphabet == 32:
                    pass
                else:
                    random_file.write(chr(random_alphabet))
            random_file.close()
        print(random_file.closed, count, 32 in alphabets)
        count = self.random_get_occurence()
        y_freq = []
        y_freq_log = []
        normal_tuple = count.most_common()
        for element in normal_tuple:
            y_freq_log.append(math.log(element[1]))
            y_freq.append(element[1])
        x_rank_log = [math.log(i) for i in range(1, len(y_freq) + 1)]
        x_rank = [i for i in range(1, len(y_freq) + 1)]
        plt.subplot(1, 2, 1)
        plt.plot(x_rank, y_freq, 'bo', markersize=1)
        plt.title("Random Corpus Normal Graph", fontsize = 18)
        plt.ylabel('Frequency', fontsize=15)
        plt.xlabel('Rank', fontsize=15)
        plt.xticks(rotation = 45)
        plt.subplot(1, 2, 2)
        plt.plot(x_rank_log, y_freq_log, 'bo', markersize=1)
        plt.title("Random Corpus Log-Log Graph", fontsize = 18)
        plt.ylabel('Frequency', fontsize=15)
        plt.xlabel('Rank', fontsize=15)
        plt.xticks(rotation = 45)
        plt.suptitle("Random Corpus Graph", y=1.05, fontsize=24)
        plt.show()
        return random_corpus
    
    def random_corpus_token_type(self):    
        """
            Analyze the token size and the vocabulary size of the the randomly generated corpus.
            Plot the results in the normal and the log-log format.
        """
        plt.figure(figsize=(20, 10))
        x_token_size = []
        y_vocab_size = []
        x_token_sizelog = []
        y_vocab_sizelog = []
        token_size = 0
        vocab_size = 0
        token = []
        with open(self.file, mode = "r", encoding="utf-8") as random_file:
            for line in random_file:
                words = line.split()
                for word in words:
                    if word in token:
                        pass
                    else:
                        vocab_size += 1
                        token.append(word)
                    token_size += 1
                    if token_size % 5000 == 0:
                        x_token_size.append(token_size)
                        y_vocab_size.append(vocab_size)
                        x_token_sizelog.append(math.log(token_size))
                        y_vocab_sizelog.append(math.log(vocab_size))
            random_file.close()
        print(random_file.closed)
        plt.subplot(1, 2, 1)
        plt.plot(x_token_size, y_vocab_size, 'bo', markersize=1)
        plt.title("Random Corpus Normal Graph", fontsize = 18)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.xticks(rotation = 45)
        plt.subplot(1, 2, 2)
        plt.plot(x_token_sizelog, y_vocab_sizelog, 'bo', markersize=1)
        plt.title("Random Corpus Log-Log Graph", fontsize = 18)
        plt.ylabel('Vocabulary Size', fontsize=15)
        plt.xlabel('Token Size', fontsize=15)
        plt.xticks(rotation = 45)
        plt.suptitle("Random Corpus Graph", y=1.05, fontsize=24)
        plt.show()
```
